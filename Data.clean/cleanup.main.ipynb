```python
df_education_quits = pd.read_excel("Education&Health.Quits.xlsx")
df_financial_quits = pd.read_excel("./Financial.Quits.xlsx")
df_government_quits = pd.read_excel("./Government.Quits.xlsx")
df_manufacturing_quits = pd.read_excel("Manufactuing.Quits.xlsx")
df_trade_quits = pd.read_excel("./Trade.tran.U.Quits.xlsx")
```

```python
df_manufacturing_quits['Yearly Total'] = df_manufacturing_quits.sum(axis=1)
manufacturing_totals_df = df_manufacturing_quits.loc[:, ['Year', 'Yearly Total']]

manufacturing_totals_df.head(5)
```

```python
# Calculate Yearly Total for each DataFrame
df_education_quits['Yearly Total'] = df_education_quits.sum(axis=1) - df_education_quits['Year']
df_financial_quits['Yearly Total'] = df_financial_quits.sum(axis=1) - df_financial_quits['Year']
df_government_quits['Yearly Total'] = df_government_quits.sum(axis=1) - df_government_quits['Year']
df_manufacturing_quits['Yearly Total'] = df_manufacturing_quits.sum(axis=1) - df_manufacturing_quits['Year']
df_trade_quits['Yearly Total'] = df_trade_quits.sum(axis=1) - df_trade_quits['Year']

# Create yearly totals
yearly_totals_education = df_education_quits.loc[:, ['Year', 'Yearly Total']]
yearly_totals_financial = df_financial_quits.loc[:, ['Year', 'Yearly Total']]
yearly_totals_government = df_government_quits.loc[:, ['Year', 'Yearly Total']]
yearly_totals_manufacturing = df_manufacturing_quits.loc[:, ['Year', 'Yearly Total']]
yearly_totals_trade = df_trade_quits.loc[:, ['Year', 'Yearly Total']]
```

```python
# Create a list of DataFrames
dfs = [
    (yearly_totals_education, 'Education & Health Services'),
    (yearly_totals_financial, 'Financial Activities'),
    (yearly_totals_government, 'Government'),
    (yearly_totals_manufacturing, 'Manufacturing'),
    (yearly_totals_trade, 'Trade, Transport & Utilities')
]

# Combine DataFrames
combined_df = pd.concat([df.assign(Industry=industry) for df, industry in dfs], axis=0)

# Reset the index
combined_df.reset_index(drop=True, inplace=True)

combined_df = combined_df.pivot_table(index='Industry', columns='Year', values='Yearly Total', aggfunc='sum')
combined_df.reset_index(inplace=True)

combined_df.head()
```
